```python
def convertToDict(x):
    splited_items = [a.split('=') for a in x]
    n=0
    for i in splited_items:
        if len(i) == 1:
            n+=1
            i.insert(0, n)
    for i in splited_items:
        i[1].replace(",","]")
    for i in range(len(splited_items)):
        splited_items[i][1]=(splited_items[i][1]).replace("[","")
        splited_items[i][1]=(splited_items[i][1]).replace("]","")
        splited_items[i][1]=(splited_items[i][1]).replace(",","")
    dictionary = {}
    for item in splited_items:
        key = item[0]
        value = item[1]
        dictionary[key] = value
    return dictionary
```

```python
s = s.apply(lambda x: convertToDict(x))
s
```

```python
new_df = s.apply(pd.Series)
new_df
```
